```python
import anvil.server
from PIL import Image
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
from io import BytesIO

# Load the model (ensure the path is correct!)
#model_path = '/content/drive/MyDrive/Project/Code/your_model.h5'
#model = load_model(model_path)
model_path = '/content/drive/MyDrive/Project/Code/keras_models/mobilenetv2_skin_classification_model.keras'
# Load the model
model = tf.keras.models.load_model(model_path)
resize = 224
# --- PROCESS IMAGE ---
@anvil.server.callable
def load_and_process_image(file):
    """Processes the uploaded image and returns the predicted skin type."""
    try:
        # Read image from Anvil upload
        image = Image.open(BytesIO(file.get_bytes()))
        image = image.resize((224, 224))
        image_array = img_to_array(image)
        image_array = np.expand_dims(image_array, axis=0)
        image_array = preprocess_input(image_array)

        # Make prediction
        predictions = model.predict(image_array)
        print(f"✅ Predictions: {predictions}")
        predicted_class_index = np.argmax(predictions)
        print(f"✅ Predicted Class Index: {predicted_class_index}")
        class_names = ['acne', 'dry', 'oil']
        predicted_class = class_names[predicted_class_index]

        print(f"✅ Predicted Class: {predicted_class}")
        return predicted_class

    except Exception as e:
        print(f"❌ Error processing image: {e}")
        return "error"

# --- GET RECOMMENDATIONS ---
if not hasattr(anvil.server, 'get_recommendations'):
  @anvil.server.callable
  def get_recommendations(predicted_class):
        # Function code...


    """Fetches product recommendations based on the predicted skin type."""
    try:
        # Load recommendations CSV
        recommendations_df = pd.read_csv('/content/drive/MyDrive/Project/Code/img.csv')

        # Filter recommendations based on predicted skin type (case-insensitive)
        predicted_class_lower = str(predicted_class).lower()
        filtered_recommendations = recommendations_df[
            recommendations_df['Skin Type'].str.lower() == predicted_class_lower
        ]

        # Return recommendations as a list of dictionaries
        if not filtered_recommendations.empty:
            recommendations_list = [
                {
                    "product": row["Product"],
                    "recommendation": row["Recommendation"],
                    "website": row["Website"],
                     "link": row["Link"],
                     "image": row["Image Link"]
                }
                for _, row in filtered_recommendations.iterrows()
            ]
            print(f"✅ Recommendations found: {len(recommendations_list)} items")
            return recommendations_list

        else:
            print("❌ No recommendations found for this skin type.")
            return [{"product": "No recommendations found"}]

    except Exception as e:
        print(f"❌ Error loading recommendations: {e}")
        return [{"product": "Error fetching recommendations"}]


```
